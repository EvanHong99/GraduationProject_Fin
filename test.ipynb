## 3维PCA展示

```python
reduced=PCA(n_components=3).fit_transform(all_change)
print(reduced.shape)
plt.plot(np.sum(reduced,axis=1))
plt.plot(all_change.mean(axis=1).values)
plt.show()
```

```python
coords=np.matmul(reduced.T,all_change.values).T
```

```python
from sklearn.cluster import OPTICS
tags=OPTICS(min_samples=5).fit_predict(coords)
print(tags)
```

```python
%matplotlib notebook

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(
    coords[:, 0],
    coords[:, 1],
    coords[:, 2],
    c=tags+5,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

## 高维pca，实际运算

```python
pca=PCA(n_components=10)
pca.fit(all_change)
reduced=pca.transform(all_change)
print(reduced.shape)
plt.plot(np.sum(reduced,axis=1))
plt.plot(all_change.mean(axis=1).values)
plt.show()
```
