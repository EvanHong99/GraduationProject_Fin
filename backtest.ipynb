```python
import backtrader as bt
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline

class ConstSpread(bt.Strategy):
    params = (("printlog", False), ("quantity", 1000))

    def log(self, txt, dt=None, doprint=False):
        """Logging function for strategy"""
        if self.params.printlog or doprint:
            dt = dt or self.datas[0].datetime.date(0)
            print(f"{dt.isoformat()}, {txt}")

    def __init__(self):
        self.delta = 0.0001
        self.Vw = self.delta / (1 - self.delta) * np.eye(2)
        self.Ve = 0.001

        self.beta = np.zeros(2)
        self.P = np.zeros((2, 2))
        self.R = np.zeros((2, 2))

        self.position_type = None  # long or short
        self.quantity = self.params.quantity

        # mycode
        self.scalar=2#用来设置close头寸的时机

    def next(self):

        x = np.asarray([self.data0[0], 1.0]).reshape((1, 2))
        y = self.data1[0]


        self.R = self.P + self.Vw  # state covariance prediction
        yhat = x.dot(self.beta)  # measurement prediction

        Q = x.dot(self.R).dot(x.T) + self.Ve  # measurement variance

        e = y - yhat  # 价差measurement prediction error

        K = self.R.dot(x.T) / Q  # Kalman gain

        self.beta += K.flatten() * e  # State update
        self.P = self.R - K * x.dot(self.R)

        sqrt_Q = np.sqrt(Q)

        # print(e)
        # print(sqrt_Q)

        if self.position:
            if self.position_type == "long" and e > -sqrt_Q:
                self.close(self.data0)
                self.close(self.data1)
                self.position_type = None
            elif self.position_type == "short" and e < sqrt_Q:
                self.close(self.data0)
                self.close(self.data1)
                self.position_type = None
            # 真的是那个alpha吗
            self.log(f"beta: {self.beta[0]}, alpha: {self.beta[1]}",doprint=True)


        else:
            if e < -self.scalar*sqrt_Q:
                self.sell(data=self.data0, size=(self.quantity * self.beta[0]))
                self.buy(data=self.data1, size=self.quantity)

                self.position_type = "long"
            elif e > self.scalar*sqrt_Q:
                self.buy(data=self.data0, size=(self.quantity * self.beta[0]))
                self.sell(data=self.data1, size=self.quantity)
                self.position_type = "short"

        self.log(f"beta: {self.beta[0]}, alpha: {self.beta[1]}",doprint=False)





class KalmanPair(bt.Strategy):
    params = (("printlog", False), ("quantity", 1000))

    def log(self, txt, dt=None, doprint=False):
        """Logging function for strategy"""
        if self.params.printlog or doprint:
            dt = dt or self.datas[0].datetime.date(0)
            print(f"{dt.isoformat()}, {txt}")

    def __init__(self):
        self.delta = 0.0001
        self.Vw = self.delta / (1 - self.delta) * np.eye(2)
        self.Ve = 0.001

        self.beta = np.zeros(2)
        self.P = np.zeros((2, 2))
        self.R = np.zeros((2, 2))

        self.position_type = None  # long or short
        self.quantity = self.params.quantity

        # mycode
        self.scalar=2#用来设置close头寸的时机

    def next(self):

        x = np.asarray([self.data0[0], 1.0]).reshape((1, 2))
        y = self.data1[0]

        self.R = self.P + self.Vw  # state covariance prediction
        yhat = x.dot(self.beta)  # measurement prediction

        Q = x.dot(self.R).dot(x.T) + self.Ve  # measurement variance

        e = y - yhat  # 价差measurement prediction error

        K = self.R.dot(x.T) / Q  # Kalman gain

        self.beta += K.flatten() * e  # State update
        self.P = self.R - K * x.dot(self.R)

        sqrt_Q = np.sqrt(Q)

        # print(e)
        # print(sqrt_Q)

        if self.position:
            if self.position_type == "long" and e > -sqrt_Q:
                self.close(self.data0)
                self.close(self.data1)
                self.position_type = None
            elif self.position_type == "short" and e < sqrt_Q:
                self.close(self.data0)
                self.close(self.data1)
                self.position_type = None
            self.log(f"beta: {self.beta[0]}, alpha: {self.beta[1]}",doprint=True)


        else:
            if e < -self.scalar*sqrt_Q:
                self.sell(data=self.data0, size=(self.quantity * self.beta[0]))
                self.buy(data=self.data1, size=self.quantity)

                self.position_type = "long"
            elif e > self.scalar*sqrt_Q:
                self.buy(data=self.data0, size=(self.quantity * self.beta[0]))
                self.sell(data=self.data1, size=self.quantity)
                self.position_type = "short"

        self.log(f"beta: {self.beta[0]}, alpha: {self.beta[1]}",doprint=False)

class MLStrategy(bt.Strategy):
    params = (("printlog", False), ("quantity", 1000))

    def log(self, txt, dt=None, doprint=False):
        """Logging function for strategy"""
        if self.params.printlog or doprint:
            dt = dt or self.datas[0].datetime.date(0)
            print(f"{dt.isoformat()}, {txt}")

    def __init__(self):
        self.delta = 0.0001
        self.Vw = self.delta / (1 - self.delta) * np.eye(2)
        self.Ve = 0.001

        self.beta = np.zeros(2)
        self.P = np.zeros((2, 2))
        self.R = np.zeros((2, 2))

        self.position_type = None  # long or short
        self.quantity = self.params.quantity

        # mycode
        self.scalar=2#用来设置close头寸的时机

        self.iteration=0
        self.history=pd.DataFrame(columns=['ds','y'])
        self.index=0

    def next(self):

        x = np.asarray([self.data0[0], 1.0]).reshape((1, 2))
        print(f"x {self.data0[0]}")
        y = self.data1[0]
        print(f"y {self.data1[0]}")

        print(self.datas[1].close)
        self.history=self.history.append(pd.DataFrame({'ds':self.datas[1].datetime.date(0),'y':self.datas[1].close[-1]},index=[self.index]))
        print(self.history)
        if self.index==10:
            assert 0
        self.R = self.P + self.Vw  # state covariance prediction
        yhat = x.dot(self.beta)  # measurement prediction

        Q = x.dot(self.R).dot(x.T) + self.Ve  # measurement variance

        e = y - yhat  # 价差measurement prediction error

        K = self.R.dot(x.T) / Q  # Kalman gain

        self.beta += K.flatten() * e  # State update
        self.P = self.R - K * x.dot(self.R)

        sqrt_Q = np.sqrt(Q)

        # print(e)
        # print(sqrt_Q)

        if self.position:
            if self.position_type == "long" and e > -sqrt_Q:
                self.close(self.data0)
                self.close(self.data1)
                self.position_type = None
            elif self.position_type == "short" and e < sqrt_Q:
                self.close(self.data0)
                self.close(self.data1)
                self.position_type = None
            # 真的是那个alpha吗
            self.log(f"beta: {self.beta[0]}, alpha: {self.beta[1]}",doprint=True)


        else:
            if e < -self.scalar*sqrt_Q:
                self.sell(data=self.data0, size=(self.quantity * self.beta[0]))
                self.buy(data=self.data1, size=self.quantity)

                self.position_type = "long"
            elif e > self.scalar*sqrt_Q:
                self.buy(data=self.data0, size=(self.quantity * self.beta[0]))
                self.sell(data=self.data1, size=self.quantity)
                self.position_type = "short"

        self.log(f"beta: {self.beta[0]}, alpha: {self.beta[1]}",doprint=False)
        self.index+=1

def min_max_norm(df):
    return (df-df.min())/(df.max()-df.min())

def run():

    # my data
    startdate = datetime.datetime(2021,  1, 1)
    enddate = datetime.datetime(2021,12, 31)

    pairs=[('601398.XSHG', '600000.XSHG'),
            # ('601398.XSHG', '600050.XSHG'),
            # ('601398.XSHG', '601818.XSHG'),
            # ('600000.XSHG', '600050.XSHG')
            ]

    for p in pairs:
        cerebro = bt.Cerebro(stdstats=True)
        cerebro.addstrategy(MLStrategy)
        # cerebro.addstrategy(KalmanPair)
        # cerebro.addstrategy(TestStrategy)
        
        testdata=jqdatasdk.get_price(p[0],start_date=startdate,end_date=enddate,frequency='daily',fields=["open"	,"high"	,"low"	,"close"	,"volume"],panel =False)
        testdata1=jqdatasdk.get_price(p[1],start_date=startdate,end_date=enddate,frequency='daily',fields=["open"	,"high"	,"low"	,"close"	,"volume"],panel =False)

        dataframe = testdata.reset_index().rename(columns={"index":"Date"	,"open":"Open"	,"close":"Close"	,"high":"High"	,"low":"Low"	,"volume":"Volume"}).set_index("Date")
        dataframe1 = testdata1.reset_index().rename(columns={"index":"Date"	,"open":"Open"	,"close":"Close"	,"high":"High"	,"low":"Low"	,"volume":"Volume"}).set_index("Date")
        

        # print('--------------------------------------------------')
        print(dataframe1.head()) 
        print(dataframe1.tail()) 
        plt.plot(min_max_norm(dataframe['Close']))
        plt.plot(min_max_norm(dataframe1['Close']))
        plt.show()
        # print('--------------------------------------------------')

        # Pass it to the backtrader datafeed and add it to the cerebro
        data = bt.feeds.PandasData(dataname=dataframe)
        data1 = bt.feeds.PandasData(dataname=dataframe1)

        cerebro.adddata(data)
        cerebro.adddata(data1)

        cerebro.broker.setcommission(commission=0.0001)
        cerebro.broker.setcash(100_000.0)

        print(f"Starting Portfolio Value: {cerebro.broker.getvalue():.2f}")
        cerebro.run()
        print(f"Final Portfolio Value: {cerebro.broker.getvalue():.2f}")
        cerebro.plot(iplot=True)


run()
```
